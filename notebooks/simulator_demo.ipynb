# Simulator demo

```python
from common import Config, State, generate_pyplot_animation
from simulator import CartPoleSimulator

from IPython.display import HTML
```

```python
import matplotlib.pyplot as plt

init = State(
    cart_position=0.0,
    cart_velocity=0.0,
    pole_angle=3.0,
    pole_angular_velocity=0.0
)

config = Config()
simulator = CartPoleSimulator()
simulator.reset_to(config, init)

states = []
timestamps = []

t = 0.0
step = 0.05 # seconds
step_n = 100

for _ in range(step_n):
    states.append(simulator.get_state())
    timestamps.append(t)
    simulator.advance(t)
    t += step
```

```python
anim = generate_pyplot_animation(config, states, timestamps=timestamps)
HTML(anim.to_jshtml())
```
